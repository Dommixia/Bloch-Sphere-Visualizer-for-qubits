### Important Libraries

```python
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np
```

### Representing the qubit
here the qubit has not been affected by any gate, so it stays at the top


```python
qc = QuantumCircuit(1)
state = Statevector.from_instruction(qc)
fig = plot_bloch_multivector(state)
```

### X Gate (Quantum Not Gate)
The X gate flips the qubit completely. Rotates the Bloch sphere arrow 180° around the X axis, moving it from the north pole to the south pole.

```python
qcx = QuantumCircuit(1)
qc.x(0)
state = Statevector.from_instruction(qc)
fig = plot_bloch_multivector(state)
```

### Y GATE
Flips the qubit like X but adds a phase factor. Rotates 180° around the Y axis.

```python
qc = QuantumCircuit(1)

qc.y(0)
state = Statevector.from_instruction(qc)
fig = plot_bloch_multivector(state)
```

### Z Gate
Leaves |0⟩ unchanged but flips the phase of |1⟩. Rotates 180° around the Z axis. No visible effect when starting from |0⟩.

```python
qc = QuantumCircuit(1)

qc.z(0)
state = Statevector.from_instruction(qc)
fig = plot_bloch_multivector(state)
```

### Hadmard Gate
Creates superposition. Puts the qubit exactly halfway between |0⟩ and |1⟩. Moves the arrow from the north pole to the equator. The most important gate in quantum computing

```python
qc = QuantumCircuit(1)

qc.h(0)
state = Statevector.from_instruction(qc)
fig = plot_bloch_multivector(state)
```

### S Gate
Rotates the arrow 90° around the Z axis. Applies a phase shift of 90° (π/2) to |1⟩. Only visible effect when qubit is already on the equator.


```python
qc = QuantumCircuit(1)

qc.s(0)
state = Statevector.from_instruction(qc)
fig = plot_bloch_multivector(state)
```

### RY Gate
Rotates the arrow by a custom angle around the Y axis. Most commonly used in QML as it keeps amplitudes real.

```python
qc = QuantumCircuit(1)
qc.ry(np.pi/2, 0)
state = Statevector.from_instruction(qc)
fig = plot_bloch_multivector(state)
```
